# Benfords Law And It's Application with Election Data

## History:

The discovery of Benford's law goes back to 1881, when the Canadian-American astronomer Simon Newcomb noticed that in logarithm tables the earlier pages (that started with 1) were much more worn than the other pages. Newcomb's published result is the first known instance of this observation and includes a distribution on the second digit as well. Newcomb proposed a law that the probability of a single number N being the first digit of a number was equal to log(N + 1) − log(N).

The phenomenon was again noted in 1938 by the physicist Frank Benford,who tested it on data from 20 different domains and was credited for it. His data set included the surface areas of 335 rivers, the sizes of 3259 US populations, 104 physical constants, 1800 molecular weights, 5000 entries from a mathematical handbook, 308 numbers contained in an issue of Reader's Digest, the street addresses of the first 342 persons listed in American Men of Science and 418 death rates. The total number of observations used in the paper was 20,229. This discovery was later named after Benford (making it an example of Stigler's law). 

## What is Benford’s Law?

Benford’s Law is an observation about the frequency distribution of leading digits in many real-life sets of numerical data. The Law states that in many naturally occurring collections of numbers, the leading significant digit is likely to be small. For example, in sets that obey the Law, the number 1 appears as the most significant digit about 30% of the time, while 9 appears as the most significant digit less than 5% of the time. If the digits were distributed uniformly, they would each occur about 11.1% of the time. Benford’s Law also makes predictions about the distribution of second digits, third digits, digit combinations, and so on.

### The graph below shows how the distribution of Benford’s Law look.

```python
import numpy as np
import matplotlib.pyplot as plt
num_1_to_10 =list(range(1,10))
benford_law = [np.log10(1+1/i)*100 for i in num_1_to_10]
plt.bar(num_1_to_10,benford_law)
plt.xticks(ticks=num_1_to_10)
plt.title("Benfods Law Graph")
plt.xlabel("Leading Digits")
plt.ylabel("Frequency")
plt.show()
```

### Scientifically, Benford's Law is based on base-10 logarithms that show the probability that the leading digit of a number will be n can be calculated as ***log10(1+1/n)***

## Application 

### Election data

Walter Mebane, a political scientist and statistician at the University of Michigan, was the first to apply the second-digit Benford's law-test (2BL-test) in election forensics. Such analysis is considered a simple, though not foolproof, method of identifying irregularities in election results.Scientific consensus to support the applicability of Benford's law to elections has not been reached in the literature. A 2011 study by the political scientists Joseph Deckert, Mikhail Myagkov, and Peter C. Ordeshook argued that Benford's law is problematic and misleading as a statistical indicator of election fraud.Their method was criticized by Mebane in a response, though he agreed that there are many caveats to the application of Benford's law to election data




## How is Benford’s Law applied to the real world?

Benford’s law has many great applications in exploratory data analysis. To give an example, forensic accountants typically use Benford’s Law to screen for irregularities in financial reports. Financial datasets tend to follow Benford’s law unless human intervention is involved, which can sometimes lead to fraud. I want to emphasize that variances from Benford’s Law do not prove fraud but rather indicate human intervention. Additional research will most likely be needed to understand the nature of the human intervention.

## DataFrame

```python
# Data downloaded as excel from https://chicagoelections.gov/en/election-results-specifics.asp
# and saved as csv to dataexport.csv

data = pd.read_csv('chicago_dataexport.csv', skiprows=10)
data = data.dropna()
data = data[~(data['Precinct'].isin(['Total', 'Precinct']))]
data.reset_index(inplace=True)
data
```

## Applying Benford's Law on Chicago Election 2018(Using Python)



### The following graphs show the results of a Benford's law test on the number of votes cast for each candidate in the 2018 Chicago election. The blue line shows the expected distribution of first digits according to Benford's law, while the red line shows the actual distribution of first digits. As you can see, the actual distribution deviates from the expected distribution, which suggests that there may have been some irregularities in the election.

```python
data =data[['Joseph R. Biden & Kamala D. Harris ',
       'Donald J. Trump & Michael R. Pence ', 
       'Howie Hawkins & Angela Walker ', 
       'Gloria La Riva & Leonard Peltier ', 
       'Brian Carroll & Amar Patel ', 
       "Jo Jorgensen & Jeremy ''Spike'' Cohen "]]
# removing end indention of column name
data.columns=data.columns.str.replace(' ','_')
data.columns=[i[:-1] for i in data.columns]


```

```python
df=pd.DataFrame()
df.index=range(1,10)
df['Benfords_Percentage']=[numpy.log10(1+1/i) for i in range(1,10)]
loop2=[]
for i in data.columns:
   df[i+'_count'] = data[i].str[0].astype('Int32').value_counts().sort_index().iloc[1:]
   loop2.append(i+'_count')
df.fillna(0,inplace=True)
for i in loop2:
   df[i+'Percentage']=[x/df[i].sum() for x in df[i]]


figure,ax=plt.subplots(nrows=2,ncols=3,figsize=(17,15))
loop=list(ax[0])+list(ax[1])
candidates=list(data.columns)


def drawing_chart(ax,candidate):
    ax.plot(df.index,df['Benfords_Percentage'], marker='o', color='red', label='Expected (Benford\'s law)')
    ax.bar(df.index,df[candidate+'_countPercentage'],label='Percentage')
    ax.legend()
    ax.set_xticks(ticks=list(df.index[1:]))
    ax.set_title(candidate)
    ax.set_xlabel("Leading Digit")
    ax.set_ylabel("Frequency")
    

for i in range(len(loop)):
   drawing_chart(loop[i],candidates[i])

plt.legend()
plt.suptitle("Chicago", fontsize=18)
plt.show()
```

## Conclusion

Benford's law is a statistical law that states that the first digit of a number in a large data set will be 1 about 30% of the time, 2 about 17% of the time, and so on. This law has been applied to a variety of data sets, including voting data. In the case of the 2018 Chicago election, Benford's law was applied to the number of votes cast for each candidate. The results showed that the first digits of the votes cast for the two main candidates, Rahm Emanuel and Jesus "Chuy" Garcia, were not distributed in accordance with Benford's law. This suggests that there may have been some irregularities in the election.

However, it is important to note that Benford's law is not a perfect test for fraud. There are a number of factors that can cause the first digits of a data set to deviate from Benford's law, including rounding errors, data entry errors, and even natural variations in the data. As a result, it is important to consider other factors when evaluating the results of a Benford's law test.

In the case of the 2018 Chicago election, there are a number of other factors that could have contributed to the deviation from Benford's law. For example, the election was held during a time of high voter turnout, which can lead to more rounding errors. Additionally, the election was contested by two well-known candidates, which could have led to more natural variations in the data.

As a result, it is difficult to say definitively whether or not there was fraud in the 2018 Chicago election. However, the deviation from Benford's law does raise some concerns about the integrity of the election. More investigation is needed to determine whether or not any fraud actually took place.
